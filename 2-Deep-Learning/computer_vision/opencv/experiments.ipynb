## Samplewise_center
Calc mean for each sample individualy

```python
import tensorflow
import numpy as np
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.mnist.load_data()
```

```python
x_train[0]
```

```python
x_train[0].shape
```

```python
for i in range(len(x_train)):
  mean = np.mean(x_train[i], axis=(0, 1))
  x_train[i] = x_train[i] - mean
```

```python
x_train[0]
```

# featurewise_std_normalization

Normalization by std: divide the inputs by std of the dataset

```python
import tensorflow
import numpy as np
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.mnist.load_data()
```

```python
std = np.std(x_train, axis= (0,1))
x_train = (x_train)/std
```

```python
x_train[0]
```

# samplewise_std_normalization

```python
import tensorflow
import numpy as np
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.mnist.load_data()
```

```python
for i in range(len(x_train)):
  std = np.std(x_train[i], axis= (0,1))
  x_train[i] = x_train[i]/std
```

# CLAHE
(Contrast Limited Adaptive Histogram Equalization)

```python
img = cv2.imread('/content/example-image.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
clahe = cv2.createCLAHE(clipLimit=80.0, tileGridSize=(8, 8))
```

```python
clahe_image = clahe.apply(img)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(clahe_image)
```

## Brightness

```python
image = cv2.imread('/content/example-image.jpg')
alpha = 1.5 # > 1 more bright and < 1 less bright
beta = 0 # Const value that is added after multiply pixels for aplha
brightened_image = cv2.addWeighted(image, alpha, np.zeros_like(image), 0, beta)
from google.colab.patches import cv2_imshow
cv2_imshow(brightened_image)
```

# Histogram Equalization

```python
image = cv2.imread('/content/example-image.jpg')
image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
# Equalize channel Y (luminance)
image_yuv[:,:,0] = cv2.equalizeHist(image_yuv[:,:,0])

# back to  BGR
equalized_image = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR)

from google.colab.patches import cv2_imshow
cv2_imshow(equalized_image)
```

## Blur

```python
image = cv2.imread('/content/example-image.jpg')
```

```python
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
from google.colab.patches import cv2_imshow
cv2_imshow(gaussian_blur)
```

```python
median_blur = cv2.medianBlur(image, 5)
from google.colab.patches import cv2_imshow
cv2_imshow(median_blur)
```

```python
bilateral_blur = cv2.bilateralFilter(image, 9, 75, 75)
from google.colab.patches import cv2_imshow
cv2_imshow(bilateral_blur)
```
